## Visualize Satellite Transfer

**Description :**
- `true_anomaly` : Determines the initial position of Earth and Mars (in radians).
- `epoch` : Number of days from the `ture_anomaly` position after which the satellite will be launched from Earth
- `dt` : Simulation time-step. Keep it small for higher accuracy (in days).
- `duration` : The length of simulation (in days).
- `C3` : 
    - Characteristic energy of the satellite with which it is escaping the Earths gravitational pull (in $km^2/s^2$).
    - Change this to observe different satellite trajectories.
- `speed` : 
    - The number of days that are skipped for each plot update for visualization.
    - Change this to make the simulation go faster.

**Legend for the Visualization Plot :**
- Sun : Yellow Dot
- Earth : Green Dot
- Mars : Red Dot
- Satellite : Blue Asterisk


```python
# import libraries
import numpy as np
from utils import visualise_satellite

# initialize parameters
true_anomaly = np.array([0.0, np.pi/3]) # initial position of planets
epoch   = 10        # duration by which to advance the initial planets position (days)
dt      = 0.01      # time-step (days)
C3      = 3.46**2  # characteristic energy (km^2/s^2)
duration= 500       # simulation duration (days)
speed   = 300       # simulation visualization speed

# run simulation
visualise_satellite(epoch, dt, C3, duration, speed, true_anomaly)
```
